```python
import pandas as pd
data = pd.read_csv('LoansTrainingSet.csv', low_memory = False)
```

```python
data.head()
```

# Incorrect Formats

```python
data['Years in current job'].unique()
```

```python
data['Years in current job'].replace(['10+ years', '4 years', '6 years', '5 years', 'n/a', '3 years',
       '2 years', '< 1 year', '1 year', '7 years', '9 years', '8 years'], [10, 4, 6, 5, 5, 3, 2, 0, 1, 7, 9, 8], inplace = True)
```

```python
data['Home Ownership'].replace('HaveMortgage', 'Home Mortgage', inplace = True)
```

```python
data['Purpose'].replace('other', 'Other', inplace = True)
```

```python
data['Monthly Debt'] = data['Monthly Debt'].str.replace('$', '')
data['Monthly Debt'] = data['Monthly Debt'].str.replace(',', '')
data['Monthly Debt'] = data['Monthly Debt'].astype(float)
```

```python
data['Maximum Open Credit'].replace('#VALUE!', 0, inplace = True)
data['Maximum Open Credit'] = data['Maximum Open Credit'].astype(int)
```

# Deal with Duplicates

```python
dups = data.duplicated('Loan ID')
```

```python
dups[dups == True].head(10)
```

```python
data.ix[6:8]
```

6 is the original, 7 is the duplicate & what is missing is Credit Score & Annual Income.

```python
data.ix[12:16]
```

Current Loan Amount : Take the max because that maximizes our risk. <br>
Annual Income : Minimum, because it tells the lowest they can afford to pay back. <br>
Monthly Debt : Maximum, because that assess how much they have in total without taking into account the new loan. <br>
Current Credit Balance : Maximum, this is how much they owe in all, so we want to have the most to consider what the customer owes the most. <br>
Maximum Open Credit : Minimum, smallest amount assess risk further.

Credit Score : Minimum, because we want to assess the more risk.

```python
maximums = data[['Loan ID', 'Current Loan Amount', 'Monthly Debt', 'Current Credit Balance']].groupby('Loan ID').max()
```

```python
maximums.head()
```

```python
minimums = data[['Loan ID', 'Credit Score', 'Annual Income', 'Maximum Open Credit']].groupby('Loan ID').min()
```

```python
minimums.head()
```

```python
data.columns
```

```python
## WILL TAKE TIMES. PLURAL TIME. TO RUN.
rest = data[['Loan ID', 'Loan Status', 'Term', 'Years in current job', 'Home Ownership', 'Purpose', 'Years of Credit History', 'Bankruptcies', 'Months since last delinquent', 'Number of Open Accounts', 'Number of Credit Problems', 'Tax Liens']].groupby('Loan ID').max()
```

```python
maximum = maximums.reset_index()
minimums = minimums.reset_index()
rest = rest.reset_index()
```

```python
maximum.head()
```

```python
cleaned = pd.merge(maximum, minimums)
cleaned = pd.merge(cleaned, rest)
```
